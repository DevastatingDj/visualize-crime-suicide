```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

crime_df = pd.read_csv('../Datasets/rajanand-crime-in-india/01_District_wise_crimes_committed_IPC_2001_2012.csv')
crime_df

crime_df_d = crime_df[crime_df['STATE/UT']=='DELHI UT']
crime_df_d = crime_df_d.groupby('YEAR').sum()
crimes = crime_df_d['TOTAL IPC CRIMES']


suicides_df = pd.read_csv('../Datasets/Suicides_in_India.csv')
suicides_df.head()

suicides_df_d = suicides_df[suicides_df['State']=='DELHI (UT)']
suicides_df_d = suicides_df_d.groupby('Year').sum(axis=1)
suicides_count = suicides_df_d.Total



data = pd.DataFrame([crimes, suicides_count])
data = data.T
data.corr()

fig = plt.figure(figsize=(11, 7), dpi=400)
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.09, 0.5, 0.35, 0.3])
ax1.plot(suicides_count, label = 'Suicide Committed')
ax2.plot(crimes, label = 'Crimes Committed')
ax1.set( title='NEW DELHI, NATIONAL CAPITAL TERRITORY OF DELHI, INDIA ')
ax2.set(title="Number of crimes committed (over the period 2001-2012)")
fig.suptitle("Number of suicides committed (over the period 2001-2012)")
labels = [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012]
ax1.set_xticks(labels, minor=True)
ax1.set_xticklabels(labels, fontdict=None, minor=True)
fig.savefig('Trends1.png')


```

# Similarity

### From the year 2001 to 2002, there is a decrease in both the crime rate and suicide rate

### From the year 2003 to 2006, there is a increase in both the crime rate and suicide rate

### From the year 2006 to 2008, there is a decrease in both the crime rate and suicide rate

### From the year 2008, the suicide count and crimes committed both go on to increase rapidly
